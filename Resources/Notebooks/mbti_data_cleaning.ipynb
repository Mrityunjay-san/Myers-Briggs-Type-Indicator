```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/drive/MyDrive/mbti_1.csv")
df
```

```python
type_count = df.groupby('type').agg({'type':'count'})
type_count
```

```python
type_chart = type_count.plot(kind='bar')

type_chart.set_xlabel("Personality Types")
type_chart.set_ylabel("Number of Samples")

plt.show()
```

```python
new_df = df.copy()
new_df['http_per_post'] = df['posts'].apply(lambda x:x.count('http')/50)
new_df 
```

```python
pattern = "(http.*?\s)"
df['without_url'] = df['posts'].replace(pattern, " ", regex=True)
df
```

```python
new_df['qm_per_post'] = df['without_url'].apply(lambda x: x.count('?')/50)
new_df
```

```python
new_df['exclaim_per_post'] = df['without_url'].apply(lambda x:x.count('!')/50);
new_df
```

```python
pattern_string = "(\|\|\|)"
df['plain_text'] =  df['without_url'].replace(pattern_string, " ", regex=True)
df
```

```python
new_df['length_per_post'] = df['plain_text'].apply(lambda x:len(x)/50)
new_df
```

```python
new_df['digits_per_post'] = df['plain_text'].apply(lambda x: sum(c.isdigit() for c in x)/50)
new_df
```

```python
eda_analysis = new_df.groupby('type').mean()
eda_analysis
```

```python
eda_analysis.plot(kind = 'bar', subplots=True, title="Writing style analysis (per post)", layout=(5, 1), sharex=True, sharey=False, legend=False, figsize=(8,12), rot=0)
```

```python
punctuation = "[^\w\s]"
df['plain_text'] = df['plain_text'].replace(punctuation, " ", regex=True)
df.head()
```

```python
underscore = '\_'
df['plain_text'] = df['plain_text'].replace(underscore, " ", regex=True)
df.head()
```

```python
numbers = '\d+'
df['plain_text'] = df['plain_text'].replace(numbers, " ", regex=True)
df.head()
```

```python
one_letter_words = '\W*\b\w\b'
df['plain_text'] = df['plain_text'].replace(one_letter_words, " ", regex=True)
df.head()
```

```python
df['plain_text'] = df['plain_text'].str.lower()
df.head()
```

```python
cleaned_df = df[['type','plain_text']]
cleaned_df.head()
```
